```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set()
```

```python
from pathlib import Path
EXP_DIR = Path('../../experiments/')
```

```python
berea_mink = pd.read_csv('../../data/berea_64/minkowski64.csv')
```

```python
logs = pd.read_csv(EXP_DIR / '2019-04-29_14_19' / 'logs.tsv', sep='\t')
```

```python
berea_mink.head()
```

```python
logs.head()
```

```python
x = logs['iter'].values
```

```python
fig = plt.figure(figsize=(16, 11))
ax = fig.gca()
ax.plot(x, logs['loss_g'], label='generator')
ax.plot(x, logs['loss_d'], label='discriminator')
ax.set_yscale('log')

for tick in ax.xaxis.get_major_ticks():
    tick.label.set_fontsize(14)

for tick in ax.yaxis.get_major_ticks():
    tick.label.set_fontsize(14)

ax.set_xlabel('', fontsize=18) 
ax.set_ylabel(' ', fontsize=18) 
ax.legend(fontsize=16)
plt.title(f"  ", fontsize=20)
plt.show()
plt.close(fig)
```

```python
fig = plt.figure(figsize=(16, 11))
ax = fig.gca()
ax.plot(x, logs['V'], label='V_GAN')

top_V = berea_mink['V'].mean() + berea_mink['V'].std()
bot_V = berea_mink['V'].mean() - berea_mink['V'].std()

ax.axhspan(ymin=bot_V, ymax=top_V, alpha=0.1, color='green', hatch='/', label='training')

for tick in ax.xaxis.get_major_ticks():
    tick.label.set_fontsize(14)

for tick in ax.yaxis.get_major_ticks():
    tick.label.set_fontsize(14)

ax.set_xlabel('', fontsize=18) 
ax.set_ylabel('$V$', fontsize=18) 
ax.legend(fontsize=16)
plt.title(f"  $V$   ", fontsize=20)
plt.show()
plt.close(fig)
```
